## Getting Breast Cancer data

Using load_breast_cancer from sklearn.datasets as per the documentation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer

df = load_breast_cancer(as_frame = True)
```

# Correlations

Identifying Correlations

```python
corr_matrix = df.frame.corr()
corr_matrix["target"].sort_values(ascending=False)
```

# Preprocessing

Preprocessing the data

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
  ('scaler', StandardScaler()),
])
```

```python
df_std = num_pipeline.fit_transform(df.data)
```

```python
from sklearn.model_selection import train_test_split

# Split into validation set and training set
X_train, X_test, y_train, y_test = train_test_split(df_std, df.target, test_size=0.2, random_state=1)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# KNN Models

```python
# KNN Model
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score

knn_cls = KNeighborsClassifier(n_neighbors=7)
knn_cls.fit(X_train, y_train)
```

```python
# Accuracy Using Cross-Validation
knn_cls_scores = cross_val_score(knn_cls, X_train, y_train, scoring="neg_mean_squared_error", cv=5)
knn_cls_rmse_scores = np.sqrt(-knn_cls_scores)
knn_cls_rmse_scores
```

```python
# Prediction (Evaluating with test-set)
pred_knn=knn_cls.predict(X_test)

print("Predictions:", list(pred_knn))
print("Labels     :", list(y_test))
```

```python
# Confusion matrix for KNN
conf_mx_knn = confusion_matrix(y_test, pred_knn)
conf_mx_knn
```

```python
# Accuracy
accuracy_knn = knn_cls.score(X_test, y_test)
accuracy_knn
```

```python
# Precision Score
precision_score(y_test, pred_knn, average = "weighted" )
```

```python
# Recall Score
recall_score(y_test, pred_knn, average = "weighted")
```

```python
# f1 score
f1_score(y_test, pred_knn, average= "weighted")
```

# Logistic Regession Model

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score

softmax_reg = LogisticRegression(solver="lbfgs", multi_class="multinomial", C=10)
softmax_reg.fit(X_train, y_train)
```

```python
# Accuracy Using Cross-Validation
softmax_reg_scores = cross_val_score(softmax_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=5)
softmax_reg_rmse_scores = np.sqrt(-softmax_reg_scores)
softmax_reg_rmse_scores
```

```python
# Predictions (Evaluating with test-set)
pred_softmax=softmax_reg.predict(X_test)
print("Predictions:", list(pred_softmax))
print("Labels     :", list(y_test))
```

```python
# Confusion matrix for Logistic Regression
conf_mx_softmax = confusion_matrix(y_test, pred_softmax)
conf_mx_softmax
```

```python
# Accuracy
accuracy_softmax = softmax_reg.score(X_test, y_test)
accuracy_softmax
```

```python
# Precision Score
print(precision_score(y_test, pred_softmax, average = None))
precision_score(y_test, pred_softmax, average = "weighted" )
```

```python
# Recall Score
recall_score(y_test, pred_softmax, average = None)
recall_score(y_test, pred_softmax, average = "weighted")
```

```python
# f1 Score
f1_score(y_test, pred_softmax, average= None)
f1_score(y_test, pred_softmax, average= "weighted")
```
